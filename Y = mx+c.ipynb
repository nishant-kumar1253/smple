```python
def hypothesis(X, theta):
    y_ = theta[0] + theta[i]*X 
    return y_
```

```python
def cost(X, Y, theta):
    m = X.shape[0]
    total_loss = 0
    for i in range(m):
        y_predt = hypothesis(X[i], theta)
        y_gt = Y[i]
        total_loss+= (y_predt - y_gt)**2
        
        return total_loss/m # calculate avg.
```

```python
def gradient(X, Y, theta):
    m = X.shape[0]
    grad = np.zeros((2,), int)
    
    for i in range(m):
        y_predt = hypothesis(X[i], theta)
        y_gt = Y[i]
        
        grad[0]+= (y_predt - y_gt)*1 #x[0]=1
        grad[1]+= (y_predt - y_gt)*X[1]       # put formula of gradient
        
        return grad/m
```

```python
def gradient_descent(X, Y, learning_rate= 0.001, epochs = 100):
    
    theta = np.zeros((2,), int) # intilization theta
    cost_epoch =[]
    
     # measure how it run and updated by 100 
    
    for i in range(epochs):
        lost = cost(X, Y, theta)
        grad = gradient(X, Y, theta)
        cost_epoch.append(loss)
        
        # now intilize gradient descent formula
        theta[0] = theta[0]-learning_rate*grad[0]
        theta[1] = theta[1]-learning_rate*grad[1]
        
        return theta, cost_epoch  
```
